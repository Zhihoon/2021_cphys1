```python
import numpy as np
import matplotlib.pyplot as plt
import random

x=[random.randint(0,10) for i in range(10)]
y=[random.randint(0,10) for i in range(10)]
plt.plot(x,y,'o')
```

```python
x=[10,9,8,6,3,3,2,0,1,2]
y=[5,8,9,7,6,4,4,3,9,8]
plt.plot(x,y,'*-')
```

```python
def dist(x,y):
    z=0
    for i in range(9):
        z+=np.sqrt((x[i+1]-x[i])**2+(y[i+1]-y[i])**2)
    return z
```

```python
dist(x,y)
```
