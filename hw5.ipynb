```python
import math
import matplotlib.pyplot as plt
import numpy as np
from scipy import linalg
```

```python
#d=2*k-m*w^2 이고 k=1, m=1 이므로 d=2-w^2이다.
#그리고 d<=2이다.(w는 실수)
```

```python
#1-1.b_1=C=1, b_2~40=0"
M0=np.ones((39,39))
zh0=np.diag(M0[0],k=0)
zh1=(-1)*np.diag(M0[0],k=1)
zh2=(-1)*np.diag(M0[0],k=-1)
zh1+zh2
```

```python
def Mk(d):
    return np.diag([d-1]+[d for i in range(38)]+[d-1])
    while d<=2:
        return Mk(d)
Mk(1) #d=1
```

```python
def M(d):
    return zh1+zh2+Mk(d)
M(1)
```

```python
b=[1]+[0 for i in range(39)]
```

```python
np.linalg.solve(M(1),b)
```

```python
plt.plot([i for i in range(40)],np.linalg.solve(M(1),b))
plt.legend("")
plt.xlabel('$i$')
plt.ylabel('$a_i$')
plt.show()
print("방정식의 해의 값이 주기적으로 나온다.(d=1)")
```

```python
plt.plot([i for i in range(40)],np.linalg.solve(M(math.sqrt(2)),b))
plt.legend("")
plt.xlabel('$i$')
plt.ylabel('$a_i$')
plt.show()
print("방정식의 해의 값이 주기적으로 나온다.(d=루트2)")
```

```python
plt.plot([i for i in range(40)],np.linalg.solve(M(math.sqrt(0.5)),b))
plt.legend("")
plt.xlabel('$i$')
plt.ylabel('$a_i$')
plt.show()
print("방정식의 해의 값이 주기적으로 나온다.(d=루트0.5)")
```

```python
#2.b_1=b_40=C=1, b_2~39=0"
B=[1]+[0 for i in range(38)]+[1]
np.linalg.solve(M(1),B)
```

```python
plt.plot([i for i in range(40)],np.linalg.solve(M(1),B))
plt.legend('')
plt.xlabel('$i$')
plt.ylabel('$a_i$')
plt.show()
print("방정식의 해의 값이 주기적으로 나온다.(d=1)")
```
